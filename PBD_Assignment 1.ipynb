# b.1) Plotting on piechart

```python
plt.pyplot.pie(df['state'].value_counts(),explode=[0,0.1,0.1],shadow=True,labels=['Malignant','Benign','Non-nodule'],
               colors=['white','blue','green']
               ,autopct='%1.1f%%'
               ,startangle=90)
```

```python
import matplotlib.pyplot as plt
import numpy as np
labels = 'Malignant','Benign','Non-nodule'
sizes = np.array([100, 54, 93])
colors = ['yellow', 'gold', 'blue']

def absolute_value(val):
    a  = np.round(val/100.*sizes.sum(), 0)
    return a

plt.pie(sizes, labels=labels, colors=colors,
        autopct=absolute_value, shadow=True)
```

```python
import matplotlib.pyplot as plt
import numpy as np
labels = 'Male','Female'
sizes = np.array([119, 128])
colors = ['pink','cyan']

def absolute_value(val):
    a  = np.round(val/100.*sizes.sum())
    return a

plt.pie(sizes, labels=labels, colors=colors,
        autopct=absolute_value, shadow=True)
```
